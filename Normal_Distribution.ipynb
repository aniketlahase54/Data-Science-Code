```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
beml_df = pd.read_csv("/content/BEML.csv")
beml_df.head()
```

```python
glaxo_df = pd.read_csv("/content/GLAXO.csv")
glaxo_df.head()
```

```python
beml_df = beml_df[['Date','Close']]
glaxo_df = glaxo_df[['Date','Close']]
```

```python
beml_df
```

```python
glaxo_df
```

```python
beml_df.dtypes
```

```python
glaxo_df.dtypes
```

```python
beml_df = beml_df.set_index(pd.DatetimeIndex(beml_df["Date"]))
```

```python
glaxo_df = glaxo_df.set_index(pd.DatetimeIndex(glaxo_df["Date"]))
```

```python
beml_df.head()
```

```python
glaxo_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.plot(glaxo_df.Close)
plt.xlabel('Time')
plt.ylabel('Close Price')
```

```python
plt.plot(beml_df.Close)
plt.xlabel('Time')
plt.ylabel('Close Price')
```

```python
beml_df["gain"] = beml_df.Close.pct_change(periods = 1)
glaxo_df["gain"] = glaxo_df.Close.pct_change(periods = 1)
```

```python
beml_df
```

```python
glaxo_df
```

```python
beml_df = beml_df.dropna()
glaxo_df = glaxo_df.dropna()
```

```python
beml_df
```

```python
glaxo_df
```

```python
plt.figure(figsize = (8,6))
plt.plot(beml_df.gain)
plt.xlabel('Time')
plt.ylabel('gain')
```

```python
plt.figure(figsize = (8,6))
plt.plot(glaxo_df.gain)
plt.xlabel('Time')
plt.ylabel('gain')
```

```python
plt.figure(figsize = (10,6))
plt.suptitle("Distribution", fontsize=25)
plt.subplot(1,2,1)
sns.distplot(beml_df.gain, label ="Beml")
plt.xlabel("gain")
plt.ylabel("Density")
plt.legend()

plt.subplot(1,2,2)
sns.distplot(glaxo_df.gain, label ="Glaxo")
plt.xlabel("gain")
plt.ylabel("Density")
plt.legend()
```

```python
plt.figure(figsize = (10,6))
sns.distplot(beml_df.gain, label ="Beml")
sns.distplot(glaxo_df.gain, label ="Glaxo")
plt.xlabel("gain")
plt.ylabel("Density")
plt.legend()
```

```python
glaxo_df_ci = stats.norm.interval(0.95,
loc = glaxo_df.gain.mean(),
scale = glaxo_df.gain.std())
print("Mean at 95% confidence interval is:",np.round(glaxo_df_ci,4))
```

```python
beml_df_ci = stats.norm.interval(0.95,
loc = beml_df.gain.mean(),
scale = beml_df.gain.std())
print("Mean at 95% confidence interval is:",np.round(beml_df_ci,4))
```
