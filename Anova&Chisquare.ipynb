```python
group1_setosa = df[df["target"]==0]["sepal length (cm)"]
group2_versicolor = df[df["target"]==1]["sepal length (cm)"]
group3_virginica = df[df["target"]==2]["sepal length (cm)"]
```

```python
group1_setosa.mean(), group2_versicolor.mean(), group3_virginica.mean()
```

```python
import seaborn as sns
sns.displot(data = df,
             x="sepal width (cm)",
             hue="target",
             fill="True",
             kind="kde",
             palette=["Blue","Orange","Green"])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,5))
plt.scatter(range(1,51),group1_setosa)
plt.scatter(range(1,51),group2_versicolor)
plt.scatter(range(1,51),group3_virginica)

plt.legend(["setosa","versicolor","virginica"])
plt.show()
```

```python
from scipy import stats
stats.f_oneway(group1_setosa,group2_versicolor,group3_virginica)
```

```python
df1 = pd.DataFrame(iris.data)
stats.f_oneway(df1.iloc[:,0],df1.iloc[:,1],df1.iloc[:,2],df1.iloc[:,3])
```

```python
from scipy.stats import chi2_contingency
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data = [[14,4],
        [0,10]]
data
```

```python
df = pd.DataFrame(data,
                  index=["Athlete","Non-Athlete"],
                  columns=["Non-smoker","Smoker"])
```

```python
df
```

```python
df.plot(kind ="bar",stacked=False)
```

```python
df.plot(kind ="bar",stacked=True)
```

```python
chi2,p,dof,exp = chi2_contingency(data)
```

```python
chi2_contingency(data)
```

```python
chi2
```

```python
p
```

```python
dof
```

```python
exp
```

```python
a=0.05
if p<a:
  print("Reject null hypothesis.Smoking and being Athlete are dependent on each other")
else:
  print("Accept null hypothesis.Smoking and Athlete are not dependent on each other")
```
