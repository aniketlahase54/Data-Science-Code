```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams["figure.figsize"] = (10,5)
plt.rcParams["figure.dpi"] = 250
sns.set_style("darkgrid")
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("/content/Wholesale.csv")
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc= StandardScaler()
```

```python
x =  sc.fit_transform(df)
```

```python
x
```

```python
from sklearn.cluster import DBSCAN
```

```python
db =  DBSCAN(eps=0.5, min_samples=3)
```

```python
y = db.fit_predict(x)
y
```

```python
x
```

```python
plt.scatter(x[:,0], x[:,1], c=y, cmap='rainbow')
```

```python
from sklearn.metrics import silhouette_score

```

```python
silhouette_score(x,y)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
neighbor = NearestNeighbors(n_neighbors=3).fit(x)
```

```python
d,index = neighbor.kneighbors(x)
```

```python
neighbor.kneighbors(x)
```

```python
d
```

```python
index
```

```python
d[:,1]
```

```python
distance = np.sort(d[:,1])
```

```python
distance
```

```python
plt.plot(distance)
plt.axhline(y=2,linestyle ="--", color="red")
plt.xlabel("Index")
plt.ylabel("EPS")
plt.title("K-Distance Gragh")
```

```python
db =  DBSCAN(eps=2, min_samples=3)
yP = db.fit_predict(x)
yP
```

```python
plt.scatter(x[:,0], x[:,1], c=yP, cmap='viridis')
```

```python
silhouette_score(x,yP)
```
