```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv("/content/retail_dataset.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
items = (df["0"].unique())
items
```

```python
encode_vals = []
for index, row in df.iterrows():
    labels = {}
    uncommons = list(set(items) - set(row))

    commons = list(set(items).intersection(row))
    for uc in uncommons:
      labels[uc] = 0
    for com in commons:
      labels[com] = 1
    encode_vals.append(labels)
```

```python
encode_vals
```

```python
ohe_df = pd.DataFrame(encode_vals)
```

```python
ohe_df
```

```python
freq_items = apriori(ohe_df, min_support=0.2, use_colnames=True)
```

```python
freq_items
```

```python
asso_rules = association_rules(freq_items, metric="confidence", min_threshold=0.6)
```

```python
asso_rules
```

```python
asso_rules.sort_values("confidence",ascending=False)
```
