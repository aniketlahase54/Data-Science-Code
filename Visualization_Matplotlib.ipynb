```python
from math import *
plt.figure(figsize=(16,9))
x = list (range(-10,11))
y = [i**2 for i in x]
plt.subplot(3,4,1)
plt.plot(x,y)

y = [i**3 for i in x]
plt.subplot(3,4,8)
plt.plot(x,y)

y = [sin(i) for i in x]
plt.subplot(3,4,10)
plt.plot(x,y)
```
